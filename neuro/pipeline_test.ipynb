```python
from collections import namedtuple
from pipeline import Pipeline
margs = namedtuple('args', 'env arena_config num_episodes')
env_path = '../env/macro2'
ac = "configurations/curriculum4/25.yml"
# ac = '../competition_configurations/5-10-1.yml'
args = margs(env=env_path, arena_config=ac, num_episodes=10)
pip = Pipeline(args)
```

```python
x = pip.run()
```

```python
df = x['t']['objects']
df1 = df[ 'visible']
df1 = df1[df1!=-1].index.tolist()
visible = ""
for i in df1:
    visible += f"visible({round(i*-1)})\n"
visible
```

```python
from clyngor import ASP

lp = """                

present(398).
present(X):- visible(X).
obstructs(X,Y) :- present(X), present(Y), visible(X), not visible(Y).
initiate(explore(X,Y)) :- visible(X), obstructs(X,Y).
initiate(interact(398)):- visible(398).
initiate(rotate(X)):-not initiate(X).

check(visible(Y)):- initiate(explore(X,Y)).
check(time):- initiate(explore(X,Y)).
check(time):- initiate(interact(398)).

"""
answer_sets = ASP(lp)
answer_sets = list(list(answer_sets)[0])
answer_sets
```
