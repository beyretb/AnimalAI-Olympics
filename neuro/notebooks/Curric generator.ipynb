```python
from random import randrange
locs = {
    'agent_x':0,
    'agent_z':0,
    'agent_rot':0,
    'goal_x':0,
    'goal_z':0,
    'wall_x':0,
    'wall_z':0,
    'wall_size':0
}
def valid_loc(diff=0):
    res = locs.copy()
    agent_x = randrange(1,39)
    agent_z = randrange(1,10-diff)
    if diff==0:
        wall_x = agent_x
    else:
        wall_x = randrange(10+diff,30-diff)
    wall_z = agent_z + randrange(6,7+diff*2)
    if wall_z>39:
        wall_z = 33
    wall_size = randrange(2+diff*3,3+diff*5)
    if (wall_x + wall_size) > 39:
        wall_size = 39-wall_x
    if (wall_x-wall_size)<0:
        wall_size = int(wall_x/2-1)
    goal_x = wall_x
    goal_z = wall_z + randrange(3,10)
    if goal_z>39:
        goal_z = 39
    wall_diff = wall_x - agent_x
    if wall_diff>0: # wall is to the right
        agent_rot = int(60*wall_diff/40)
    else: # wall is to the left
        agent_rot = int(360+(60*wall_diff/40))
        
    for k in locs:
        locs[k] = locals()[k]
    return locs
num_arenas = 50
curric_num = 7
for i in range(0,num_arenas):
    
    locs = valid_loc(int(i/10))
    if any((v>39)|(v<0) for k,v in locs.items() if k!='agent_rot'):
        raise Exception(f"BAD ENTRY {locs}")
    arena = """!ArenaConfig
    arenas:
      -1: !Arena
        items:
        - !Item
          name: Agent
          positions:
          - !Vector3 {{x: {agent_x}, y: 0, z: {agent_z}}}
          rotations: [{agent_rot}]

        - !Item
          name: GoodGoal
          positions:
          - !Vector3 {{x: {goal_x}, y: 0, z: {goal_z}}}
          sizes:
          - !Vector3 {{x: 1, y: 1, z: 1}}
        - !Item
          name: Wall
          positions:
          - !Vector3 {{x: {wall_x}, y: 0, z: {wall_z}}}
          sizes:
          - !Vector3 {{x: {wall_size}, y: 5, z: 2}}
          rotations: [0]
          colors: 
          - !RGB {{r: 153, g: 153, b: 153}}
        pass_mark: 0
        t: 250""".format(**locs)
    text_file = open(f"configurations/curriculum{curric_num}/{i+25}.yml", "w")
    text_file.write(arena)
    text_file.close()
```

```python
a_nums = []
for i in range(num_arenas):
    a_nums.append(f"{i}.yml")
a_thresh = []
for i in range(num_arenas-1):
    a_thresh.append(f"0.65")
json = {
    "measure":"reward",
    "min_lesson_length":1,
    "signal_smoothing":False,
    "thresholds":a_thresh,
    "configuration_files":a_nums
}
text_file = open(f"configurations/curriculum{curric_num}/AnimalAI.json", "w")
text_file.write(str(json).replace("'", '"'))
text_file.close()
```
