```python
broken_up = {k:[0 for _ in range(len(results))] for k,v in mode.items()}
for seed in results:
    for arena, ares in results[seed]['Spatial Elimination'].items():
        if arena=='success_rate':
            continue
        for typ, ars in mode.items():
            if (arena in ars)&(ares['success']):
                broken_up[typ][seed]+=1
```

```python
import pandas as pd
df = pd.DataFrame(broken_up)
df['total'] = df.sum(axis=1)/27
for i in mode:
    df[i] = df[i]/len(mode[i])
df.loc['mean'] = df.mean(axis=0)
# df.describe()
df
```
